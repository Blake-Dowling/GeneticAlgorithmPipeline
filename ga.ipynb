```python
ga_param_grid = {
    'initialization_function': [initialize_random],
    'selection_function': [tournament_selection],
    'crossover_function': [linear_crossover],
    'mutation_function': [uniform_mutate],
    'population_size': [50, 100, 200],
    'crossover_rate': [0.7, 0.8, 0.9],
    'mutation_rate': [0.01, 0.05, 0.1],
    'elite_count': [1, 2]
}
def ga_scorer(estimator, X, y):
    return estimator.score(X, y)
grid_search_ga = GridSearchCV(
    estimator=GAOptimizer(
                            initialization_function=initialize_random,
                            fitness_function=de_jongs_f1, 
                            selection_function=tournament_selection,
                            crossover_function=linear_crossover,
                            mutation_function=uniform_mutate,
                            max_generations=100,
                            verbose_ga=False, 
                            n_ga_runs_per_eval=3), # Instantiate wrapper with defaults
    param_grid=ga_param_grid,
    cv=2, # Using 2-fold CV to show it runs multiple times, each fold calls 'fit'
    verbose=2, # Verbosity for GridSearchCV itself
    n_jobs=2, # Setting to 1 to see sequential output clearly for each 'fit' call
    scoring=None
)

dummy_X = np.array([[0], [1]]) # 2 samples
dummy_y = np.array([0, 0])   # 2 samples

print("--- Starting GridSearchCV to tune GA Hyperparameters ---")
# Fit GridSearchCV with dummy data
grid_search_ga.fit(dummy_X, dummy_y)
print("--- GridSearchCV for GA finished! ---")

# Get the best GA hyperparameters
print(f"\nBest GA Hyperparameters found: {grid_search_ga.best_params_}")
# The best_score_ here is the average of the scores returned by the wrapper.score() method.
print(f"Best average score (negative objective value): {grid_search_ga.best_score_:.4f}")
print(f"This corresponds to an average best objective value of: {-grid_search_ga.best_score_:.6f}")

# You can also inspect all results
import pandas as pd
results_df = pd.DataFrame(grid_search_ga.cv_results_)
# Filter for relevant columns and sort by rank
relevant_cols = [col for col in results_df.columns if col.startswith('param_') or col in ['mean_test_score', 'rank_test_score']]
results_df = results_df[relevant_cols].sort_values(by='rank_test_score')
print("\nTop 5 GA Hyperparameter Combinations:")
print(results_df.head())
```
